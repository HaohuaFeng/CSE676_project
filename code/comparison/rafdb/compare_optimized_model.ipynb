```python
import matplotlib.pyplot as plt
import sys

sys.path.append("../../")  # Add the parent directory to the Python path
import helper.utility as utility

# Model list
model_configs = [
    "custom/v6.1_Adam_[RAF(aug12)]_LR_[L2:0.01]",
    "custom/v6.1_Adam_[RAF(aug12)]_LR_[L2:0.001]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_[L2:0.01]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_[L2:0.001]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01,smooth:0.1]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01,smooth:0.01]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01,smooth:0.2]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01,smooth:0.05]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01,smooth:0.025]",
    "custom/v7.2_Adam_[RAF(aug12)]_LR_WB(A)_[L2:0.01]",
    "custom/v7.2_Adam_[RAF(AutoAug12x5),FER(AutoAug12x5)]_LR_[L2:0.01]",
    "custom/v7.2_Adam_[RAF(AutoAug12x5),FER(AutoAug12x5)]_LRWB(A)_[L2:0.01]",
    "custom/v7.2_Adam_[RAF(AutoAug12x10)]_LR_[L2:0.01]",
    "custom/v7.2_Adam_[RAF(NP03456,aug12)]_LR",
    "DCNN/v6_Adam_[RAF(aug12)]_LR",
    "DCNN/v6_Adam_[RAF(aug12)]_LR_[L2:0.01]",
    "DCNN/v6_Adam_[RAF(aug12)]_LR_[L2:0.001]",
    "DCNN/v6_Adam_[RAF(NP03456,aug12)]_LR",
]


# Define different colors for the lines
colors = [
    "b",
    "g",
    "r",
    "c",
    "m",
    "y",
    "k",
    "#FFD700",
    "#33FF57",
    "#5733FF",
    "#FF5733",
    "#FF69B4",
    "#A52A2A",
    "#A52AA2",
    "#7FFF00",
    "#7F00FF",
    "#6495ED",
    "#DC143C",
]
plt.figure(figsize=(20, 12))

for i, model_name in enumerate(model_configs):
    identifier = model_name

    # load and plot the loss history
    record_save_path = "../../model_data/" + identifier
    data = utility.read_pickle_files(record_save_path + "/val_accuracy_history.pkl")

    max_accuracy = max(data)
    max_epoch = data.index(max_accuracy) + 1

    plt.plot(
        range(1, len(data) + 1),
        data,
        label=identifier,
        color=colors[i],
    )

    # Adjust the position of the annotation based on the data
    xytext_offset = (5, 5) if i % 2 == 0 else (-15, -15)

    plt.annotate(
        f"{max_accuracy:.3f}",
        xy=(max_epoch, max_accuracy),
        xytext=xytext_offset,  
        textcoords="offset points",
        arrowprops=dict(arrowstyle="->", lw=1),
        ha="center",
        fontsize=8,
    )

plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.ylim(0.5, 0.85)
plt.title("Validation Accuracy")
plt.legend()
plt.savefig("../../model_data/compare_optimized_model/all_validation_accuracy.png")
plt.show()
```
