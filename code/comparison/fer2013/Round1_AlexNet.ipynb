## model: AlexNet_256 + Activation Function on FC (Sigmoid, Tanh, Relu) + Optimizer (Adam, SGD)

```python
# Training Loss

import matplotlib.pyplot as plt
import sys

sys.path.append("../")  # Add the parent directory to the Python path
import helper.utility as utility

# Model list
model_configs = ["Alex_256_relu_", "Alex_256_sigmoid_", "Alex_256_tanh_"]
# optimizer list
optimizer_configs = ["Adam", "SGD"]
# Define different colors for the lines
colors = ["b", "g", "r", "c", "m", "y"]
plt.figure(figsize=(10, 6))

for i, model_name in enumerate(model_configs):
    for j, optimizer_name in enumerate(optimizer_configs):
        identifier = model_name + optimizer_name

        # load and plot the loss history
        record_save_path = "../../model_data/" + identifier
        data = utility.read_pickle_files(record_save_path + "/loss_history.pkl")
        plt.plot(
            range(1, len(data) + 1),
            data,
            label=identifier,
            color=colors[i * len(optimizer_configs) + j],
        )

plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("Training Loss")
plt.legend()
plt.savefig("../../model_data/Round1/alex_256_all_loss_history.png")
plt.show()
```
